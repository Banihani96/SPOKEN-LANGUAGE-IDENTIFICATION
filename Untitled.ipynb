```python
#from google.colab import files
from glob import glob
import os
import librosa
from librosa import feature
import numpy as np
import csv


#from pydub import AudioSegment

#audio = r"C:\Users\m.banihani\Downloads"
audio  = r"C:\Users\HP\Desktop\eng"
os.chdir(audio)

audio_files = os.listdir()
i = 0
# You dont need the number of files in the folder, just iterate over them directly using:
for file in audio_files:
    #spliting the file into the name and the extension
    #name, ext = os.path.splitext(file)
    if file.endswith(".wav"):
    #if ext == ".wav":
        fn_list_i = [
            feature.chroma_stft,
            feature.spectral_centroid,
            feature.spectral_bandwidth,
            feature.spectral_rolloff,
            feature.chroma_cqt,
            feature.chroma_cens,
            feature.melspectrogram,
            feature.mfcc,
            feature.rms,
            feature.spectral_contrast,
            #feature.spectral_flatness,
            feature.poly_features,
            feature.tonnetz]

        fn_list_ii = [feature.zero_crossing_rate]

        def get_feature_vector(y,sr,name):  
            feat_vect_i = [ np.mean(funct(y,sr)) for funct in fn_list_i]
            feat_vect_ii = [ np.mean(funct(y)) for funct in fn_list_ii]

            feature_vector =   feat_vect_i + feat_vect_ii + name 
            return feature_vector

        #build the matrix with normal audios featurized
        audios_feat = []
        Clip_Name = []
        for file in audio_files:
            if file.endswith(".wav"):
                Clip_Name.append(file.rsplit('/', 1)[-1])
                y,sr = librosa.load(file,sr=None)   
                feature_vector = get_feature_vector(y, sr,[file.rsplit('/', 1)[-1]])
                audios_feat.append(feature_vector)  
                
        norm_output = 'kalan.csv'
        header =[

            'chroma_stft',
            'spectral_centroid',
            'spectral_bandwidth',
            'spectral_rolloff',
            'chroma_cqt',
            'chroma_cens',
            'melspectrogram',
            'mfcc',
            'rms',
            'spectral_contrast',
            #feature.spectral_flatness,
            'poly_features',
            'tonnetz',
            'zero_crossing_rate']
        i=0
        #WARNING : this overwrites the file each time. Be aware of this because feature extraction step takes time.
        with open(norm_output,'+w') as f:
            csv_writer = csv.writer(f, delimiter = ',')
            csv_writer.writerow(header)
            csv_writer.writerows(audios_feat)
            #csv_writer.writerows(Clip_Name[i])
            i+=1
```

```python
def create_spectrogram_test(audio_file):
    out_image_file = audio_file + '.png'
    plt.interactive(False)
    clip, sample_rate = librosa.load(audio_file, sr=None)
    fig = plt.figure(figsize=[1.2,1.2])
    ax = fig.add_subplot(111)
    ax.axes.get_xaxis().set_visible(False)
    ax.axes.get_yaxis().set_visible(False)
    ax.set_frame_on(False)
    S = librosa.feature.melspectrogram(y=clip, sr=sample_rate)
    librosa.display.specshow(librosa.power_to_db(S, ref=np.max))
    fig.savefig(out_image_file, dpi=400, bbox_inches='tight',pad_inches=0)
    plt.close()    
    fig.clf()
    plt.close(fig)
    plt.close('all')
```
